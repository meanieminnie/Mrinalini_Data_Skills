## Oil V. Gold Comparison

```python
df_raw = pd.read_csv("oil-v-gold.csv")
df_raw
```

```python
df_raw.info()
```

### 1. Drop duplicates & NAN values
### 2. Convert objeacts into floating integers
### 3. Drop column names
### 4. Rename gold & oil columns¶

```python
#  drop duplicates 
df_raw.columns
```

```python
df = df_raw[["GC1 Comdty (Gold)","Unnamed: 1","Unnamed: 4"]]
df
```

```python
# rename column headers
df.columns = ["closing_date","px_gold","px_oil"]
df
```

```python
# delete top row
# reset index
df = df.drop(0).reset_index(drop = True)
df
```

```python
df.info()
```

```python
# convert to floating point numbers for gold
df["px_gold"] = pd.to_numeric(df["px_gold"], errors = "coerce")
df.info()
```

```python
# convert to floating point numbers for oil
df["px_oil"] = pd.to_numeric(df["px_oil"], errors = "coerce")
df.info()
```

```python
# call df
df
```

```python
# create quick line graphs of oil v. gold
df.plot(x = "closing_date",
       y = ["px_gold", "px_oil"])
```

```python
# percent change calculation for gold using .iat
old_gold = df["px_gold"].iat[0]
df["gold_pct_chg"] = \
((df["px_gold"] - old_gold)/ old_gold) * 100
df
```

```python
#  percent change oil
old_oil = df["px_oil"].iat[0]
df["oil_pct_chg"] = \
((df["px_oil"] - old_oil)/ old_oil) * 100
df
```

```python
# basic line graph
df.plot(x = "closing_date",
       y = ["gold_pct_chg", "oil_pct_chg"])
```

```python
#  add a title
df.plot(x = "closing_date",
       y = ["gold_pct_chg", "oil_pct_chg"],
       title = "Oil v. Gold Over Time")
```

```python
# why is gold orange and oil blue?
df.plot(x = "closing_date",
       y = ["gold_pct_chg", "oil_pct_chg"],
       title = "Oil v. Gold Over Time",
       style = {"gold_pct_chg": "yellow",
               "oil_pct_chg": "gray"})
```
