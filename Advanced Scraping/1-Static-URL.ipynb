# What if we wanted to scrape ALL the gas lines?


```python
# capture all the biz pages
homepage_url = "https://infopost.enbridge.com/InfoPost/"
response = requests.get(homepage_url)
soup = BeautifulSoup(response.text,"html.parser")
```

```python
dropdown_list = soup.find(id = "dropdown")
dropdown_list
```

```python
# we want the things in a tags
a_tags = dropdown_list.find_all("a")
a_tags
```

```python
# get hrefs
hrefs = [a_tag.get("href") for a_tag in a_tags]
hrefs
```

```python
# we need to extract Pipe=AG etc using regular expressions. go to regex101
# Pipe: anchor ; \w captures all the letters, numbers and characters
# import regex library
import re
```

```python
# capture pattern
# using re, i use compile function to compile all the data
# r means raw string
pat = re.compile(r"Pipe=(\w+)")
```

```python
# need to pull out the unit codes form the above list 
unit_codes = []
for unit_code in hrefs:
#     print(unit_code)
    unit_codes.append(pat.findall(unit_code)[0])
    
unit_codes
```

```python
# create url start and url end
url_start = "https://infopost.enbridge.com/InfoPost/NoticesList.asp?pipe="
url_end = "&type=CRI"
```

```python
#  create links using lc
links = [f"{url_start}{unit_code}{url_end}" for unit_code in unit_codes]
links
```

```python
# need to be kind to the server, create a timer
import time
from random import randrange
```

```python
# create a separate variable to store any broken links
# i need to create these links on the fly i.e. add f string value from links
# this is because i want to create a "unit" column with the extracted unit codes in my final df
broken_url = []
df_list=[]
total_links = len(unit_codes)
counter = 1

for unit_code in unit_codes: #need to get one unit code at a time
    print(f"Scraping {counter} of {total_links}")
    counter += 1
    response = requests.get(f"{url_start}{unit_code}{url_end}") 
    try:
        data = pd.read_html(response.text)    #
        df = data[0]
        df["unit"] = unit_code
        df_list.append(df)
    except: #this means exception, which is error
        print(f"{unit_code} was busted or had no table")
        broken_url.append(f"{url_start}{unit_code}{url_end}")
    finally:
        snooze = randrange(10,15)
        print(f"Snoozing for {snooze} seconds.")
        time.sleep(snooze)

print("Done scraping all units")
```
